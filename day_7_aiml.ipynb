```python
drop_cols=["Cabin","Name","Ticket","PassengerId"]
existing_drop_cols=[col for col in drop_cols if col in df.columns]
df.drop(columns=existing_drop_cols, inplace=True)
print("Columns after dropping Unnecessary Features: ")
print(df.columns)
```

```python
if "Age" in df.columns:
  df["Age"]=df["Age"].fillna(df["Age"].median())
if "Embarked" in df.columns:
  df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode([0]))
print("Check missing values after cleaning: ")
print(df.isnull().sum())
```

```python
X=df.drop(columns=["Survived"])
Y=df["Survived"]

print("Input Features (X) Preview")
display(X.head())
print("Target Variable (Y) Preview")
display(Y.head())
```

```python
categorical_cols=[col for col in ["Sex","Embarked"] if col in X.columns]
X=pd.get_dummies(X,columns=categorical_cols,drop_first=True)

print("\nInput Features (X) after One-Hot Encoding: ")
display(X.head())
print("Columns in X after One-Hot Encoding: ")
print(X.columns)
```

```python
label_encoder=LabelEncoder()
Y_encoded=label_encoder.fit_transform(Y)

print("\nTarget Variable (Y) after Label Encoding: ")
display(Y_encoded[:10])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y_encoded,test_size=0.2,random_state=69)
print("Train-Test Split Shapes:")
print("X-train: ",X_train.shape)
print("X-test: ",X_test.shape)
print("Y-train: ",Y_train.shape)
print("Y-test: ",Y_test.shape)
```

```python
standard_scaler=StandardScaler()
X_train_standardized=standard_scaler.fit_transform(X_train)
X_test_standardized=standard_scaler.transform(X_test)

print("First Five Rows of Standardized X-train: ")
print(X_train_standardized[:5])
```

```python
minmax_scaler=MinMaxScaler()
X_train_normanlized=minmax_scaler.fit_transform(X_train)
X_test_normalized=minmax_scaler.transform(X_test)

print("First Five Rows of Normalized X-train: ")
print(X_train_normanlized[:5])
```
