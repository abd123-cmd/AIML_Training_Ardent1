```python
from sklearn.impute import SimpleImputer

# Initialize imputer objects with distinct names
cat_imputer = SimpleImputer(strategy="most_frequent")
num_imputer = SimpleImputer(strategy="mean")

# Assuming 'df', 'cat_col', and 'cat_int' are already defined from previous cells
# Impute categorical columns
df[cat_col] = cat_imputer.fit_transform(df[cat_col])

# Impute numerical columns (using 'cat_int' which contains numerical column names)
df[cat_int] = num_imputer.fit_transform(df[cat_int])

# Display the updated DataFrame
df
```
